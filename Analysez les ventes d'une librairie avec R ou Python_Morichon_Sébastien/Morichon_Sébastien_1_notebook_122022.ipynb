### Nous allons ajouter quelques colonnes pour approndir l'analyse
- Date Fixe
- Nombre de vente total par client (en nombre)
- Nombre de vente mensuelle (nombre de vente / 23) on retire le mois d'octobre
- Nombre de vente total par client - CA (en ‚Ç¨)
- Panier moyen (en nombre)
- Panier moyen (en ‚Ç¨)

```python
#Mise en forme de la date au format (AAAA-MM-DD)
df['date_fixe'] = df['date'].dt.date
df
```

```python
# Ajout du nombre d'achat total par client
df = df.merge(
    df.groupby('client_id').count()['date'].reset_index(),
    how="left",on="client_id")
df = df.rename(columns={"date_y":"vente_total","date_x":"date"})
df
```

```python
#Ajout d'une colonne avec le nombre de vente mensuel moyen
df["vente_mensuelle"] = df["vente_total"] / 23 #(23 mois car 2ans - le mois d'octobre 2021)
df
```

```python
#Ajout d'une colonne contenant le total de CA r√©alis√© par chaque client
df = df.merge(
    df.groupby('client_id').sum()['price'].reset_index(),
    how="left",on="client_id")
df = df.rename(columns={"price_x":"price","price_y":"Total_CA"})
df
```

```python
#Ajout d'une colonne avce le panier moyen
df["panier_moyen_‚Ç¨"] = df["Total_CA"] / df["vente_total"]
df
```

```python
#Ajout d'une colonne avec les diff√©rentes classes d'√¢ge
df['classe_age'] = '18-30'
df['classe_age'].loc[df[df['√Çge']>=30].index] = '30-40'
df['classe_age'].loc[df[df['√Çge']>=40].index] = '40-50'
df['classe_age'].loc[df[df['√Çge']>=50].index] = '50-60'
df['classe_age'].loc[df[df['√Çge']>=60].index] = '60-70'
df['classe_age'].loc[df[df['√Çge']>=70].index] = '70-80'
df['classe_age'].loc[df[df['√Çge']>=80].index] = '>80'
```

## Lien entre √¢ge et cat√©gorie d'achat

L'√¢ge des clients est-il diff√©rent en fonction de la cat√©gorie d'achat ? 
- H0 : La moyenne des √¢ges des clients est identique quelque soit la cat√©gorie d'achat 
- H1 : Au moins l'une des moyennes des √¢ges des clients est diff√©rentes en fonction de la cat√©gorie d'achat 

Condition de validation du test ANOVA : 
- Type des variables (Qualitative et Quantitative) : Categ (quali) et √Çge (quanti continue)
- Independence : On utilise la fonction sample() avec 1500 entr√©es
- Normalit√© : On v√©rifie la normalit√© de la variable √Çge. Pour cela, on utilise le test de Shapiro
- Egalit√© des variances : Avec par exemple le test be bartlett

```python
#Affichage du boxplot 

plt.figure(figsize=(10,6))
plt.title("Moyenne des √¢ges en fonction de la cat√©gorie")
sns.boxplot(data=sample_df, x="√Çge", y="categ", orient="h",showfliers=False, showmeans=True)
plt.show()
```

Moyenne proche entre les cat√©gories 0 et 1

Dans un test ANOVA, on cherche √† d√©terminer si les moyennes sont significativement diff√©rentes.

H0 : Les moyennes de chaque groupe sont √©gales si p-value > 5% </br>
H1 : Les moyennes de chaque groupe ne sont pas toutes √©gales si p-value < 5%

```python
# Test de normalit√©
stat, p = shapiro(sample_df["√Çge"])
if p > 0.05:
    print("H0 : Les donn√©es ne semblent pas respecter une loi normale")
    print("p-value :",p)
else:
    print("H1 : Les donn√©es semblent respecter une loie normale")
    print("p-value :",p)
```

```python
# Test des variances
sample_df.groupby("categ")['√Çge'].agg('var')
```

```python
#Test de bartlett, test sur la variance

from scipy.stats import bartlett

stat, p = bartlett(sample_df.√Çge[sample_df.categ == 0.0],
        sample_df.√Çge[sample_df.categ == 1.0],
        sample_df.√Çge[sample_df.categ == 2.0])

if p > 0.05:
    print('H0: les variances sont √©gales')
    print("p-value :",p)
else:
    print('H1: les variances ne sont pas √©gales')
    print("p-value :",p)
```

Le test de bartlett n'√©tant pas concluant, nous devons utiliser le test de Welch ANOVA

```python
# Test Welch ANOVA

stat, p = ttest_ind(sample_df["√Çge"], sample_df["categ"])

if p > 0.05:
    print("H0: les moyennes des √©chantillons sont √©gales")
    print("pvalue : ",p)
else:
    print("H1: une ou plus des moyennes des √©chantillons sont in√©gales")
    print("pvalue : ",p)
```

Pour conclure, si au moins l'une des moyennes des √©chantillons sont in√©gales, cela signifique qu'il existe une corr√©lation entre l'√¢ge du client et la cat√©gorie des livres achet√©s
